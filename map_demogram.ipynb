```python
import numpy as np
import pickle
from matplotlib import pyplot as plt
```

```python
dem=pickle.load(open('/mnt/h/data/11.pkl','rb'))
dem_hot=np.zeros([len(dem),dem[0].data.shape[0],dem[0].data.shape[1]])
for i in np.arange(len(dem)):
    x=np.asarray(dem[i].data)
    dem_hot[i,:,:]=np.sum(x[:,:,24:],axis=2)
```

```python
plt.figure(figsize=[12,12])

plt.subplot(221)
i=8
plt.title('+ '+str(i)+' mins ')
plt.imshow(np.flipud(dem_hot[i,:,:]-dem_hot[i-1,:,:]),vmax=1e21,vmin=-1e21)
plt.subplot(222)
i=11
plt.title('+ '+str(i)+' mins ')
plt.imshow(np.flipud(dem_hot[i,:,:]-dem_hot[i-1,:,:]),vmax=1e21,vmin=-1e21)
plt.subplot(223)
i=13
plt.title('+ '+str(i)+' mins ')
plt.imshow(np.flipud(dem_hot[i,:,:]-dem_hot[i-1,:,:]),vmax=1e21,vmin=-1e21)
plt.subplot(224)
i=16
plt.title('+ '+str(i)+' mins ')
plt.imshow(np.flipud(dem_hot[i,:,:]-dem_hot[i-1,:,:]),vmax=1e21,vmin=-1e21)


```
